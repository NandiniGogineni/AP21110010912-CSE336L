```python
#Q1 Create a figure object called fig using plt.figure() , Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
import matplotlib.pyplot as plt
import pandas as pd
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
x = [1, 3, 5, 7, 9]
y = [2, 4, 6, 8, 10]
ax.plot(x, y)

ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')
ax.set_title('Plot on Axis')
plt.show()
```

```python
#Q3 Write a Python program to draw a line with suitable label in the x axis, y axis and a title.

import matplotlib.pyplot as plt
X = range(1, 50)
Y = [value * 3 for value in X]
print("Values of X:")
print(*range(1,50)) 
print("Values of Y (thrice of X):")
print(Y)
plt.plot(X, Y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Draw a line.')
plt.show()
```

```python
#Q4 Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.

import matplotlib.pyplot as plt
with open("matplotlib.txt") as f:
    data = f.read()
data = data.split('\n')
x = [row.split(' ')[0] for row in data]
y = [row.split(' ')[1] for row in data]
plt.plot(x, y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Sample graph!')
plt.show()
```

```python
#Q5 Write a Python program to plot two or more lines with legends, different widths and colors.

import matplotlib.pyplot as plt


x_values = [1, 2, 3, 4, 5]
y1_values = [1, 3, 5, 7, 9]
y2_values = [13, 20, 14, 28, 16]

plt.plot(x_values, y1_values, label='Line 1', color='black', linewidth=2)
plt.plot(x_values, y2_values, label='Line 2', color='pink', linewidth=3)
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Lines Plot')
plt.show()
```

```python
#Q7 Create plot using subplots(), scatter, and some other methods on the Axes object:

import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
fig, axs = plt.subplots(2, 1, figsize=(8, 6))
axs[0].plot(x, y1, color='blue', label='Sin(x)')
axs[0].set_title('Plot of Sin(x)')
axs[0].legend()
axs[1].plot(x, y2, color='red', label='Cos(x)')
axs[1].set_title('Plot of Cos(x)')
axs[1].legend()

plt.tight_layout()
plt.show()
```

```python
#Q8 Generate random data for x and y and create a scatter plot using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
num_points = 50
x = np.random.rand(num_points) * 10
y = np.random.rand(num_points) * 20

plt.scatter(x, y, color='green', marker='o', label='Random Data')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Random Data')

plt.legend()

plt.show()
```

```python
#Q9 Create a bar chart to compare the sales of different products. You can use fictional data or come up with your own dataset.

import numpy as np
import matplotlib.pyplot as plt

products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales = [200, 400, 600, 300, 800]

plt.bar(products, sales, color='lightpink')

plt.xlabel('Products')
plt.ylabel('Sales (units)')
plt.title('Product Sales Comparison')

plt.show()
```

```python
#Q10 Generate a random dataset and create a histogram using Matplotlib to visualize the distribution of the data.

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(54)  
data = np.random.normal(loc=0, scale=1, size=1000)  

plt.hist(data, bins=10, edgecolor='White') 
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')


plt.show()
```
